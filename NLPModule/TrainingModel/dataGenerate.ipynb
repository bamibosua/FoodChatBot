```python
import json
import ast

def export_annotations_only(df, output_path):
    with open(output_path, "w", encoding="utf8") as f:
        for ann in df["annotations"]:
            
            # Nếu là string dạng "(text, {'entities': [...]})"
            if isinstance(ann, str):
                try:
                    ann = ast.literal_eval(ann)   # convert string → Python object
                except:
                    pass
            
            # Ghi ra JSONL
            f.write(json.dumps(ann, ensure_ascii=False) + "\n")

    print("✔ Saved:", output_path)
    
export_annotations_only(df_labels, "annotations_only.jsonl")
```
